### 1.1.5.6.1. Geometrical Problems

$$
\text{Geometry} + \text{Linear Algebra} \rightarrow \text{Systematic Problem Solving}
$$

**Explanation:**

Geometrical problems combine the tools developed throughout this chapter ‚Äî lines, planes, projections, cross products, and dot products ‚Äî to solve practical problems in 2D and 3D.

Typical problems include: finding intersections of lines and planes, computing distances from points to planes, finding the equation of a plane through three points, and projecting vectors onto planes.

**Example:**

Each of the following notebooks tackles a specific type of geometrical problem using the techniques from Sections 5.1‚Äì5.5.

```python
import sympy as sp

point = sp.Matrix([1, 2, 3])
normal = sp.Matrix([0, 0, 1])

distance_to_xy_plane = abs(point.dot(normal)) / normal.norm()

print(f"Point: {list(point)}")
print(f"Distance to xy-plane: {distance_to_xy_plane}")
```
