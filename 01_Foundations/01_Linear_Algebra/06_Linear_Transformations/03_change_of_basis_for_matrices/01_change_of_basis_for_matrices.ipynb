### 1.1.6.3.1. Change of Basis for Matrices

$$
[T(\vec{v})]_{B_W} = {}_{B_W}[M_T]_{B_V}\, [\vec{v}]_{B_V}
$$

$$
{}_{B'}[M_T]_{B'} = {}_{B'}[\mathbb{1}]_{B} \;{}_{B}[M_T]_{B} \;{}_{B}[\mathbb{1}]_{B'}
$$

**Explanation:**

The coefficients of a matrix representation $M_T$ depend on the bases chosen for the input and output spaces. The same linear transformation $T$ corresponds to different matrices under different bases.

The notation ${}_{B_W}[M_T]_{B_V}$ explicitly shows the matrix depends on the input basis $B_V$ and the output basis $B_W$. Unless otherwise specified, the standard basis is assumed.

**Example:**

The reflection $R_x$ in the standard basis $B = \{\hat{e}_1, \hat{e}_2\}$:

$$
{}_B[M_{R_x}]_B = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}
$$

```python
import numpy as np

matrix_rx_standard = np.array([[1, 0], [0, -1]])
test_vector = np.array([3, 4])
print(f"R_x in standard basis B:\n{matrix_rx_standard}")
print(f"R_x @ {list(test_vector)} = {list(matrix_rx_standard @ test_vector)}")
```
