### 1.1.5.4.1. Vector Spaces

$$
(V, \mathbb{R}, +, \cdot)
$$

$$
W \subseteq V \text{ is a subspace if } \forall \vec{w}_1, \vec{w}_2 \in W,\; \alpha \in \mathbb{R}: \quad \alpha\,\vec{w}_1 + \vec{w}_2 \in W
$$

**Explanation:**

A vector space $V$ is a set of vectors equipped with addition and scalar multiplication that satisfy closure, associativity, commutativity, identity, and inverse properties.

A subspace $W$ of $V$ is a subset that is itself a vector space — it must contain the zero vector and be closed under addition and scalar multiplication.

**Example:**

The $xy$-plane $W = \{(x,y,0) \mid x,y \in \mathbb{R}\}$ is a subspace of $\mathbb{R}^3$. For any $\vec{w}_1 = (a,b,0)$ and $\vec{w}_2 = (c,d,0)$:

$$
\alpha\,\vec{w}_1 + \vec{w}_2 = (\alpha a + c,\; \alpha b + d,\; 0) \in W
$$

```python
import numpy as np

w1 = np.array([3, 4, 0])
w2 = np.array([1, -2, 0])
alpha = 2.5

combination = alpha * w1 + w2

print(f"w1 = {w1}")
print(f"w2 = {w2}")
print(f"{alpha}*w1 + w2 = {combination}")
print(f"z-component is zero (closure): {combination[2] == 0}")
```
