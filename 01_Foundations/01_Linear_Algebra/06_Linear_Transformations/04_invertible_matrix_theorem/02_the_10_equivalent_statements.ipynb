### 1.1.6.4.2. The 10 Equivalent Statements

$$
\begin{aligned}
&(1)\; A \text{ is invertible} &\quad &(6)\; \operatorname{rank}(A) = n \\
&(2)\; A\vec{x} = \vec{b} \text{ has exactly one solution } \forall\, \vec{b} &\quad &(7)\; \operatorname{rref}(A) = \mathbb{1}_n \\
&(3)\; \mathcal{N}(A) = \{\vec{0}\} &\quad &(8)\; A^\mathsf{T} \text{ is invertible} \\
&(4)\; A\vec{x} = \vec{0} \text{ has only } \vec{x} = \vec{0} &\quad &(9)\; \text{rows of } A \text{ form a basis for } \mathbb{R}^n \\
&(5)\; \text{columns of } A \text{ form a basis for } \mathbb{R}^n &\quad &(10)\; \det(A) \neq 0
\end{aligned}
$$

**Explanation:**

For an $n \times n$ matrix $A$, the following 10 statements are **all equivalent**:

1. $A$ is invertible
2. $A\vec{x} = \vec{b}$ has exactly one solution for each $\vec{b} \in \mathbb{R}^n$
3. $\mathcal{N}(A) = \{\vec{0}\}$
4. $A\vec{x} = \vec{0}$ has only the trivial solution
5. The columns of $A$ form a basis for $\mathbb{R}^n$ (linearly independent and span $\mathbb{R}^n$)
6. $\operatorname{rank}(A) = n$
7. $\operatorname{rref}(A) = \mathbb{1}_n$
8. $A^\mathsf{T}$ is invertible
9. The rows of $A$ form a basis for $\mathbb{R}^n$
10. $\det(A) \neq 0$

**Example:**

$A = \begin{bmatrix} 2 & 1 \\ 1 & 1 \end{bmatrix}$: all 10 statements are TRUE.

```python
import numpy as np
import sympy

matrix_a = np.array([[2, 1], [1, 1]])
print(f"A =\n{matrix_a}")
print(f"\n(1)  A⁻¹ =\n{np.linalg.inv(matrix_a)}")
print(f"(2)  Ax = [5,3] → x = {np.linalg.solve(matrix_a, [5, 3])}")

singular_values = np.linalg.svd(matrix_a, compute_uv=False)
nullity = np.sum(singular_values < 1e-10)
print(f"(3)  nullity = {nullity}")
print(f"(6)  rank = {np.linalg.matrix_rank(matrix_a)}")

rref_a, _ = sympy.Matrix(matrix_a).rref()
print(f"(7)  rref(A) = I: {rref_a == sympy.eye(2)}")
print(f"(8)  det(Aᵀ) = {np.linalg.det(matrix_a.T):.4f}")
print(f"(10) det(A) = {np.linalg.det(matrix_a):.4f}")
```
