### 1.1.3.5. Transpose of a Matrix

$$
\mathbf{A}^T =
\begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}^T
=
\begin{bmatrix}
a_{11} & a_{21} \\
a_{12} & a_{22}
\end{bmatrix}
$$

**Explanation:**

The transpose of a matrix flips it over its diagonal, exchanging rows with columns.
It is used in ML for switching between data representations.

**Example:**

If

$$
\mathbf{A} =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix},
$$

then

$$
\mathbf{A}^T =
\begin{bmatrix}
1 & 3 \\
2 & 4
\end{bmatrix}.
$$

```python
import numpy as np

matrix_a = np.array([[1, 2], [3, 4]])

transposed = matrix_a.T

print("A·µÄ =\n", transposed)
```
