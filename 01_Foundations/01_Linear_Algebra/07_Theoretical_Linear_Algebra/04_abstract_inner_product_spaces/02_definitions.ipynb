### 1.1.7.4.2. Definitions

$$
\begin{aligned}
&\langle \cdot, \cdot \rangle : V \times V \to \mathbb{R} \quad \text{(inner product)} \\
&\|\mathbf{u}\| \equiv \sqrt{\langle \mathbf{u}, \mathbf{u} \rangle} \quad \text{(norm)} \\
&d(\mathbf{u}, \mathbf{v}) \equiv \|\mathbf{u} - \mathbf{v}\| \quad \text{(distance)}
\end{aligned}
$$

**Explanation:**

The inner product induces two derived quantities: the norm (length) and the distance between vectors. The norm generalizes the Euclidean length, and the distance generalizes the Euclidean distance to any abstract vector space with an inner product.

**Example:**

For $\mathbf{u} = (3, 4)$:

$$
\|\mathbf{u}\| = \sqrt{\langle \mathbf{u}, \mathbf{u} \rangle} = \sqrt{9 + 16} = 5
$$

For $\mathbf{u} = (1, 0)$ and $\mathbf{v} = (0, 1)$:

$$
d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\| = \|(1, -1)\| = \sqrt{2}
$$

```python
import numpy as np

vector_u = np.array([3, 4, 1])
vector_v = np.array([1, 0, 2])

norm_u = np.sqrt(np.dot(vector_u, vector_u))
print("||u|| =", norm_u)

norm_v = np.sqrt(np.dot(vector_v, vector_v))
print("||v|| =", norm_v)

difference = vector_u - vector_v
distance_uv = np.sqrt(np.dot(difference, difference))
print("d(u, v) =", distance_uv)

print("Matches np.linalg.norm:", np.isclose(norm_u, np.linalg.norm(vector_u)))
print("Matches np.linalg.norm distance:", np.isclose(distance_uv, np.linalg.norm(difference)))
```
