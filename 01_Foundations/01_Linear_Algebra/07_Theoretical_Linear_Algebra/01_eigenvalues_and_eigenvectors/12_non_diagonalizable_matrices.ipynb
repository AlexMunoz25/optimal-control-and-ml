### 1.1.7.1.12. Non-Diagonalizable Matrices

$$
B = \begin{bmatrix} 3 & 1 \\ 0 & 3 \end{bmatrix}, \qquad \text{algebraic multiplicity} = 2, \quad \text{geometric multiplicity} = 1
$$

**Explanation:**

Not all matrices are diagonalizable. The matrix $B$ has $\lambda = 3$ as a repeated eigenvalue with algebraic multiplicity 2, but only one linearly independent eigenvector $(1, 0)^{\mathsf{T}}$, giving geometric multiplicity 1. Since we are one eigenvector short, we cannot construct a full eigenvector matrix $Q$ and $B$ is not diagonalizable.

This is called deficient geometric multiplicity: the eigenspace dimension is strictly less than the algebraic multiplicity.

**Example:**

For

$$
B = \begin{bmatrix} 3 & 1 \\ 0 & 3 \end{bmatrix}
$$

$$
B - 3I = \begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix}
$$

has $\text{rank}(B - 3I) = 1$ and $\text{nullity}(B - 3I) = 1$, confirming only one eigenvector exists.

```python
import numpy as np

matrix_b = np.array([[3, 1],
                     [0, 3]], dtype=float)

trace = matrix_b[0, 0] + matrix_b[1, 1]
determinant = matrix_b[0, 0] * matrix_b[1, 1] - matrix_b[0, 1] * matrix_b[1, 0]
eigenvalues = np.roots([1, -trace, determinant])
print("Eigenvalues:", eigenvalues)

algebraic_multiplicity = len(eigenvalues)
unique_eigenvalue = eigenvalues[0]

shifted_matrix = matrix_b - unique_eigenvalue * np.eye(2)
print("\nB - 3I:")
print(shifted_matrix)

_, singular_values, right_singular = np.linalg.svd(shifted_matrix)
null_count = np.sum(singular_values < 1e-10)
geometric_multiplicity = null_count

print("rank(B - 3I) =", 2 - geometric_multiplicity)
print("nullity(B - 3I) =", geometric_multiplicity)
print("Algebraic multiplicity =", algebraic_multiplicity)
print("Geometric multiplicity =", geometric_multiplicity)
print("Diagonalizable:", geometric_multiplicity == algebraic_multiplicity)

eigenvector = right_singular[-1]
print("\nEigenvector:", np.round(eigenvector, 4))
print("B路e_位 =", np.round(matrix_b @ eigenvector, 4))
print("位路e_位 =", np.round(unique_eigenvalue * eigenvector, 4))

print("\n--- Library method ---")
eigenvalues_lib = np.linalg.eigvals(matrix_b)
print("Eigenvalues:", eigenvalues_lib)
```
