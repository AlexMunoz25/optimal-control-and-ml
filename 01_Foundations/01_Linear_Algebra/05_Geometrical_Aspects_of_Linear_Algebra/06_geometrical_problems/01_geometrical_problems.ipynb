### 1.1.5.6.1. Geometrical Problems

$$
\text{Geometry} + \text{Linear Algebra} \rightarrow \text{Systematic Problem Solving}
$$

**Explanation:**

Geometrical problems combine the tools developed throughout this chapter — lines, planes, projections, cross products, and dot products — to solve practical problems in 2D and 3D.

Typical problems include: finding intersections of lines and planes, computing distances from points to planes, finding the equation of a plane through three points, and projecting vectors onto planes.

**Example:**

Each of the following notebooks tackles a specific type of geometrical problem using the techniques from Sections 5.1–5.5.

```python
import numpy as np

point = np.array([1, 2, 3])
normal = np.array([0, 0, 1])

distance_to_xy_plane = abs(np.dot(point, normal)) / np.linalg.norm(normal)

print(f"Point: {point}")
print(f"Distance to xy-plane: {distance_to_xy_plane}")
```
