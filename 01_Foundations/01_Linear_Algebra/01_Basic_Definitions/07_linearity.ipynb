### 1.1.1.7. Linearity

$$
f(x) =
\underbrace{\frac{a}{x}}_{\text{one-over-x}}
+ \underbrace{b}_{\text{constant}}
+ \underbrace{mx}_{\text{linear term}}
+ \underbrace{qx^2}_{\text{quadratic}}
+ \underbrace{cx^3}_{\text{cubic}}
$$

**Explanation:**

The term $mx$ is called the **linear term** because it contains $x$ raised to the first power.
All other terms are **nonlinear**, involving $x$ in the denominator or raised to powers other than one.

The linear term is unique because changes in $x$ lead to proportional changes in the output.
If $x$ doubles, the linear term doubles: $m(2x) = 2mx$.

**Example:**

For the linear function $f(x) = 3x + 2$:
- $f(1) = 5$
- $f(2) = 8$

Doubling $x$ doubles the $3x$ term ‚Äî showing perfect linear proportionality.

```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-5, 5, 200)
linear_output = 3 * x_values + 2
nonlinear_output = 3 * x_values ** 2 + 2

plt.plot(x_values, linear_output, label="Linear: f(x) = 3x + 2")
plt.plot(x_values, nonlinear_output, "--", label="Nonlinear: f(x) = 3x¬≤ + 2")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.title("Comparison: Linear vs Nonlinear Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
