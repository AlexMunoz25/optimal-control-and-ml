### 4.2 Projections


$$
\mathbf{Œ†}_P(\mathbf{u}) = \mathbf{u} - \frac{\mathbf{u} \cdot \mathbf{n}}{\|\mathbf{n}\|^2}\mathbf{n}
$$


**Explanation:**  


Projections describe how a vector decomposes into parts that are parallel and perpendicular to a given subspace.  


Given a vector $\mathbf{u}$, we can determine how much of it lies **along a direction** (projection onto a line) or **within a plane** (projection onto a plane).  




The **dot product**
$$
\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2 + u_3v_3
$$
measures alignment between two vectors and serves as the core operation for computing projections.  


Consider a plane $P$ through the origin with normal vector $\mathbf{n}$.  


The space perpendicular to this plane, denoted $P^{\perp}$, is a line in the direction of $\mathbf{n}$.  


The projection of $\mathbf{u}$ onto $P^{\perp}$ isolates the component of $\mathbf{u}$ orthogonal to the plane:  


$$
\mathbf{Œ†}_{P^{\perp}}(\mathbf{u}) = \frac{\mathbf{u} \cdot \mathbf{n}}{\|\mathbf{n}\|^2} \mathbf{n}.
$$  


To obtain the projection **onto the plane** itself, we remove this perpendicular part from $\mathbf{u}$:  


$$
\mathbf{Œ†}_P(\mathbf{u}) = \mathbf{u} - \mathbf{Œ†}_{P^{\perp}}(\mathbf{u}) 
= \mathbf{u} - \frac{\mathbf{u} \cdot \mathbf{n}}{\|\mathbf{n}\|^2}\mathbf{n}.
$$  


**Example:**  


Let  

$$
\mathbf{u} =
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix},
\quad
\mathbf{n} =
\begin{bmatrix}
1 \\ 1 \\ 1
\end{bmatrix}.
$$  


Then  

$$
\mathbf{Œ†}_P(\mathbf{u}) 
= 
\mathbf{u} - \frac{\mathbf{u} \cdot \mathbf{n}}{\|\mathbf{n}\|^2}\mathbf{n}
= 
\begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix}
-
\frac{15}{3}
\begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}
=
\begin{bmatrix} -1 \\ 0 \\ 1 \end{bmatrix}.
$$


```python
import numpy as np

u = np.array([4, 5, 6])
n = np.array([1, 1, 1])

projection_plane = u - (np.dot(u, n) / np.dot(n, n)) * n
print("Projection of u onto plane P:", projection_plane)
```
