### 1.1.6.4.4. Injective, Surjective, Bijective

$$
T \text{ is injective} \;\Leftrightarrow\; \operatorname{Ker}(T) = \{\vec{0}\}
$$

$$
T \text{ is surjective} \;\Leftrightarrow\; \operatorname{Im}(T) = \mathbb{R}^m
$$

$$
T \text{ is bijective} \;\Leftrightarrow\; T \text{ is invertible}
$$

**Explanation:**

For $T: \mathbb{R}^n \to \mathbb{R}^m$ with matrix $M_T$:

- **Injective**: $T(\vec{v}_1) \neq T(\vec{v}_2)$ for all $\vec{v}_1 \neq \vec{v}_2$ $\;\Leftrightarrow\;$ $\operatorname{Ker}(T) = \{\vec{0}\}$
- **Surjective**: $\operatorname{Im}(T) = \mathbb{R}^m$ $\;\Leftrightarrow\;$ $\mathcal{C}(M_T) = \mathbb{R}^m$
- **Bijective**: both injective and surjective $\;\Leftrightarrow\;$ $T$ is invertible

For square matrices ($T: \mathbb{R}^n \to \mathbb{R}^n$), either injective or surjective alone implies invertibility. If $T$ is injective, $\operatorname{Ker}(T) = \{\vec{0}\}$ so $T$ is invertible. If $T$ is surjective, $\operatorname{rank}(M_T) = n$ and the rank-nullity theorem gives $\mathcal{N}(M_T) = \{\vec{0}\}$.

**Example:**

- $A = \begin{bmatrix} 2 & 1 \\ 1 & 1 \end{bmatrix}$: $\operatorname{Ker} = \{\vec{0}\}$ → injective, $\operatorname{Im} = \mathbb{R}^2$ → surjective → **bijective**
- $B = \begin{bmatrix} 1 & 2 \\ 2 & 4 \end{bmatrix}$: $\operatorname{Ker} \neq \{\vec{0}\}$ → not injective, $\operatorname{rank} < 2$ → not surjective → **not bijective**

```python
import numpy as np

matrix_a = np.array([[2, 1], [1, 1]])
matrix_b = np.array([[1, 2], [2, 4]])

rank_a = np.linalg.matrix_rank(matrix_a)
rank_b = np.linalg.matrix_rank(matrix_b)
print(f"A: rank = {rank_a}, Ker = {{0}} → injective ✓, Im = R² → surjective ✓ → bijective")
print(f"B: rank = {rank_b}, Ker ≠ {{0}} → not injective, Im ≠ R² → not surjective → not bijective")
```
