### 1.1.2.1. Vector Addition

$$
\mathbf{u} + \mathbf{v} =
\begin{bmatrix}
u_1 \\
u_2 \\
\vdots \\
u_n
\end{bmatrix}
+
\begin{bmatrix}
v_1 \\
v_2 \\
\vdots \\
v_n
\end{bmatrix}
=
\begin{bmatrix}
u_1 + v_1 \\
u_2 + v_2 \\
\vdots \\
u_n + v_n
\end{bmatrix}
$$

**Explanation:**

Vector addition combines two vectors component-wise. It is commonly used in machine learning for gradient updates or geometric vector operations.

**Example:**

If

$$
\mathbf{u} = 
\begin{bmatrix}
1 \\
2
\end{bmatrix},
\quad
\mathbf{v} =
\begin{bmatrix}
3 \\
4
\end{bmatrix},
$$

then

$$
\mathbf{u} + \mathbf{v} =
\begin{bmatrix}
4 \\
6
\end{bmatrix}.
$$

```python
import numpy as np

vector_u = np.array([1, 2])
vector_v = np.array([3, 4])

result = vector_u + vector_v

print("u + v =", result)
```
