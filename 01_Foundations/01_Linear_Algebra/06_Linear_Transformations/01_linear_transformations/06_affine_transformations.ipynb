### 1.1.6.1.6. Affine Transformations

$$
\vec{y} = A(\vec{x}) = M_T \vec{x} + \vec{b}
$$

**Explanation:**

An affine transformation $A: \mathbb{R}^n \to \mathbb{R}^m$ is the combination of a linear transformation $T$ followed by a translation by a fixed vector $\vec{b}$. This is the vector generalization of the single-variable affine function $y = mx + b$.

Affine transformations are **not** linear (unless $\vec{b} = \vec{0}$) because $A(\vec{0}) = \vec{b} \neq \vec{0}$.

**Example:**

$$
A(\vec{x}) = \begin{bmatrix} 2 & 0 \\ 0 & 1 \end{bmatrix}\vec{x} + \begin{bmatrix} 1 \\ -2 \end{bmatrix}
$$

$$
A(3, 5) = \begin{bmatrix} 6 \\ 5 \end{bmatrix} + \begin{bmatrix} 1 \\ -2 \end{bmatrix} = \begin{bmatrix} 7 \\ 3 \end{bmatrix}
$$

```python
import numpy as np

stretch_matrix = np.array([[2, 0], [0, 1]])
translation = np.array([1, -2])
test_vector = np.array([3, 5])

affine_result = stretch_matrix @ test_vector + translation
print(f"A(v) = M @ v + b = {list(affine_result)}")
```
