### 1.1.6.5.1. Linear Transformations Problems

$$
M_T = \begin{bmatrix} | & | & & | \\ T(\hat{e}_1) & T(\hat{e}_2) & \cdots & T(\hat{e}_n) \\ | & | & & | \end{bmatrix}
$$

**Explanation:**

This section solves problems combining the techniques from Sections 6.1–6.4: finding matrix representations, computing image spaces and kernels, and working with function spaces.

**Example:**

The general approach: probe $T$ with the standard basis vectors, record outputs as columns of $M_T$, then analyze the resulting matrix for rank, null space, and image space.

```python
import numpy as np

matrix_t = np.array([[1, 0], [1, -1], [0, 2]])
print(f"M_T =\n{matrix_t}")
print(f"rank(M_T) = {np.linalg.matrix_rank(matrix_t)}")
print(f"Im(T) = span of {matrix_t.shape[1]} linearly independent columns")
```
