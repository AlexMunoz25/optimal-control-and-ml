### 1.1.7.7.3. Complex Matrices

$$
A \in \mathbb{C}^{m \times n}
$$

A $2 \times 2$ complex matrix has 8 parameters:

$$
\begin{bmatrix} \alpha & \beta \\ \gamma & \delta \end{bmatrix} = \begin{bmatrix} \text{Re}\{\alpha\} & \text{Re}\{\beta\} \\ \text{Re}\{\gamma\} & \text{Re}\{\delta\} \end{bmatrix} + \begin{bmatrix} \text{Im}\{\alpha\} & \text{Im}\{\beta\} \\ \text{Im}\{\gamma\} & \text{Im}\{\delta\} \end{bmatrix} i
$$

**Explanation:**

A complex matrix $A \in \mathbb{C}^{m \times n}$ has complex entries.
Each entry has real and imaginary parts, so a $2 \times 2$ complex matrix requires 8 real parameters.
Complex matrices implement linear transformations $T: \mathbb{C}^m \to \mathbb{C}^n$.

**Example:**

If

$$
A = \begin{bmatrix} 1+i & 2 \\ 3i & 4-i \end{bmatrix}, \quad \vec{v} = \begin{bmatrix} 1 \\ i \end{bmatrix},
$$

then

$$
A\vec{v} = \begin{bmatrix} (1+i)(1) + (2)(i) \\ (3i)(1) + (4-i)(i) \end{bmatrix} = \begin{bmatrix} 1 + 3i \\ 1 + 7i \end{bmatrix}.
$$

```python
import numpy as np

matrix_a = np.array([
    [1 + 1j, 2 + 0j],
    [0 + 3j, 4 - 1j]
])

print("A =")
print(matrix_a)
print("\nRe{A} =")
print(matrix_a.real)
print("\nIm{A} =")
print(matrix_a.imag)

vector_v = np.array([1 + 0j, 0 + 1j])
product = matrix_a @ vector_v

print("\nv =", vector_v)
print("Av =", product)
```
