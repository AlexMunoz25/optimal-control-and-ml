### 1.1.7.4.4. Norm

$$
\|\mathbf{u}\| = \sqrt{\langle \mathbf{u}, \mathbf{u} \rangle}
$$

$$
\begin{aligned}
&\|\mathbf{v}\| \geq 0, \quad \|\mathbf{v}\| = 0 \iff \mathbf{v} = \mathbf{0} \\
&\|k\mathbf{v}\| = |k|\|\mathbf{v}\| \\
&\|\mathbf{u} + \mathbf{v}\| \leq \|\mathbf{u}\| + \|\mathbf{v}\| \quad \text{(triangle inequality)} \\
&|\langle \mathbf{x}, \mathbf{y} \rangle| \leq \|\mathbf{x}\|\|\mathbf{y}\| \quad \text{(Cauchyâ€“Schwarz)}
\end{aligned}
$$

**Explanation:**

Every inner product induces a norm that measures the "length" of an abstract vector. All norms must satisfy non-negativity, homogeneity, the triangle inequality, and the Cauchyâ€“Schwarz inequality. Norms defined from valid inner products automatically satisfy these criteria.

**Example:**

For $\mathbf{u} = (3, 4)$:

$$
\|\mathbf{u}\| = \sqrt{9 + 16} = 5
$$

Triangle inequality:

$$
\|(1,0) + (0,1)\| = \sqrt{2} \leq 1 + 1 = 2
$$

Cauchyâ€“Schwarz:

$$
|\langle (1,2), (3,4) \rangle| = 11 \leq \sqrt{5} \cdot 5 = 5\sqrt{5} \approx 11.18
$$

```python
import numpy as np

vector_u = np.array([3, 4])
norm_u = np.linalg.norm(vector_u)
print("||u|| =", norm_u)

zero_vector = np.array([0, 0])
print("Non-negativity: ||u|| >= 0 ->", norm_u >= 0)
print("||0|| = 0 ->", np.isclose(np.linalg.norm(zero_vector), 0))

scalar_k = -3
homogeneity_left = np.linalg.norm(scalar_k * vector_u)
homogeneity_right = abs(scalar_k) * np.linalg.norm(vector_u)
print("Homogeneity: ||ku|| =", homogeneity_left, "== |k|||u|| =", homogeneity_right, "->", np.isclose(homogeneity_left, homogeneity_right))

vector_a = np.array([1, 0])
vector_b = np.array([0, 1])
triangle_left = np.linalg.norm(vector_a + vector_b)
triangle_right = np.linalg.norm(vector_a) + np.linalg.norm(vector_b)
print("Triangle inequality: ||a+b|| =", round(triangle_left, 4), "<= ||a||+||b|| =", triangle_right, "->", triangle_left <= triangle_right + 1e-10)

vector_x = np.array([1, 2])
vector_y = np.array([3, 4])
cauchy_schwarz_left = abs(np.dot(vector_x, vector_y))
cauchy_schwarz_right = np.linalg.norm(vector_x) * np.linalg.norm(vector_y)
print("Cauchy-Schwarz: |<x,y>| =", cauchy_schwarz_left, "<= ||x||||y|| =", round(cauchy_schwarz_right, 4), "->", cauchy_schwarz_left <= cauchy_schwarz_right + 1e-10)
```
