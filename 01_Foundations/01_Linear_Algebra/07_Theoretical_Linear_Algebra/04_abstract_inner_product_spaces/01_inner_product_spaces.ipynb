### 1.1.7.4.1. Inner Product Spaces

$$
\langle \cdot, \cdot \rangle : V \times V \to \mathbb{R}
$$

$$
\text{Axioms:} \quad
\begin{cases}
\text{Symmetry:} & \langle \mathbf{u}, \mathbf{v} \rangle = \langle \mathbf{v}, \mathbf{u} \rangle \\
\text{Linearity:} & \langle \alpha \mathbf{u} + \beta \mathbf{w}, \mathbf{v} \rangle = \alpha \langle \mathbf{u}, \mathbf{v} \rangle + \beta \langle \mathbf{w}, \mathbf{v} \rangle \\
\text{Positive semi-definiteness:} & \langle \mathbf{u}, \mathbf{u} \rangle \geq 0, \quad = 0 \iff \mathbf{u} = \mathbf{0}
\end{cases}
$$

$$
\|\mathbf{u}\| = \sqrt{\langle \mathbf{u}, \mathbf{u} \rangle}, \qquad d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|, \qquad \mathbf{u} \perp \mathbf{v} \iff \langle \mathbf{u}, \mathbf{v} \rangle = 0
$$

**Explanation:**

An **inner product space** is a vector space equipped with an inner product $\langle \cdot, \cdot \rangle$ satisfying symmetry, linearity, and positive semi-definiteness. These three axioms generalize the dot product to abstract spaces.

From the inner product, three derived concepts follow: the **norm** $\|\mathbf{u}\| = \sqrt{\langle \mathbf{u}, \mathbf{u} \rangle}$ (generalized length), the **distance** $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|$, and **orthogonality** $\mathbf{u} \perp \mathbf{v} \iff \langle \mathbf{u}, \mathbf{v} \rangle = 0$.

**Properties:**
- Cauchyâ€“Schwarz: $|\langle \mathbf{u}, \mathbf{v} \rangle| \leq \|\mathbf{u}\| \|\mathbf{v}\|$
- Triangle inequality: $\|\mathbf{u} + \mathbf{v}\| \leq \|\mathbf{u}\| + \|\mathbf{v}\|$
- Homogeneity: $\|\alpha \mathbf{u}\| = |\alpha| \|\mathbf{u}\|$

**Example:**

Standard dot product on $\mathbb{R}^2$: $\langle \mathbf{u}, \mathbf{v} \rangle = u_1 v_1 + u_2 v_2$

$$
\mathbf{u} = \begin{bmatrix} 3 \\ 4 \end{bmatrix}, \quad \|\mathbf{u}\| = \sqrt{9 + 16} = 5
$$

$$
\langle (1,1), (1,-1) \rangle = 1 - 1 = 0 \implies (1,1) \perp (1,-1)
$$

```python
import numpy as np

vector_u = np.array([3.0, 4.0])
vector_v = np.array([1.0, 2.0])

inner_product = np.dot(vector_u, vector_v)
norm_u = np.sqrt(np.dot(vector_u, vector_u))
distance = np.linalg.norm(vector_u - vector_v)
print(f"<u, v> = {inner_product}")
print(f"||u|| = {norm_u}")
print(f"d(u, v) = {round(distance, 4)}")

print("\n--- Axiom Verification ---")
print(f"Symmetry: <u,v>={inner_product}, <v,u>={np.dot(vector_v, vector_u)}")
alpha = 2.0
vector_w = np.array([5.0, 6.0])
left_linearity = np.dot(alpha * vector_u + vector_w, vector_v)
right_linearity = alpha * np.dot(vector_u, vector_v) + np.dot(vector_w, vector_v)
print(f"Linearity: {np.isclose(left_linearity, right_linearity)}")
print(f"Positive semi-def: <u,u>={np.dot(vector_u, vector_u)} >= 0")

print("\n--- Orthogonality ---")
orthogonal_pairs = [
    (np.array([1.0, 0.0]), np.array([0.0, 1.0])),
    (np.array([1.0, 1.0]), np.array([1.0, -1.0])),
]
for pair_a, pair_b in orthogonal_pairs:
    print(f"  <{list(pair_a)}, {list(pair_b)}> = {np.dot(pair_a, pair_b)}")

print("\n--- Cauchy-Schwarz ---")
cauchy_schwarz_holds = abs(inner_product) <= np.linalg.norm(vector_u) * np.linalg.norm(vector_v)
print(f"|<u,v>| = {abs(inner_product)} <= ||u||*||v|| = {round(norm_u * np.linalg.norm(vector_v), 4)}: {cauchy_schwarz_holds}")
```
