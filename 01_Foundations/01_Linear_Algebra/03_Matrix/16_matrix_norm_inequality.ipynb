### 1.1.3.16. Matrix Norm Inequality

$$
\|\mathbf{AB}\| \leq \|\mathbf{A}\| \cdot \|\mathbf{B}\|
$$

$$
\|\mathbf{A}\| =
\max_{\mathbf{x} \neq 0}
\frac{\|\mathbf{A}\mathbf{x}\|}{\|\mathbf{x}\|}
$$

**Explanation:**

Matrix norms satisfy the submultiplicative property: the norm of a product is at most the product of the norms.
The induced operator norm of a matrix $\mathbf{A}$ measures the maximum stretching factor of $\mathbf{A}$ acting on unit vectors.

**Example:**

For the spectral norm $\|\cdot\|_2$, derived from the Euclidean vector norm, the inequality becomes

$$
\|\mathbf{AB}\|_2 \leq \|\mathbf{A}\|_2 \|\mathbf{B}\|_2,
$$

which follows from the singular value decomposition of matrices.

```python
import sympy as sp

matrix_a = sp.Matrix([[1, 2], [3, 4]])
vector_x = sp.Matrix([1, 1])

norm_ax = (matrix_a * vector_x).norm()
norm_a_norm_x = matrix_a.norm() * vector_x.norm()

print("â€–Axâ€– =", norm_ax)
print("â€–Aâ€–â€–xâ€– =", norm_a_norm_x)
print("Inequality holds?", bool(norm_ax <= norm_a_norm_x))
```
