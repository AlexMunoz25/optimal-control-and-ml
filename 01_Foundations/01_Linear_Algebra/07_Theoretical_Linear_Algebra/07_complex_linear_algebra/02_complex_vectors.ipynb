### 1.1.7.7.2. Complex Vectors

$$
\vec{v} = (v_1, v_2, \ldots, v_n) \in \mathbb{C}^n
$$

where $v_i = a_i + b_i i$

**Explanation:**

A complex vector $\vec{v} \in \mathbb{C}^n$ is an array of $n$ complex numbers.
Each component has a real part and an imaginary part, so specifying a vector in $\mathbb{C}^n$ requires $2n$ real parameters.

**Example:**

If

$$
\vec{v} = (1 + 2i,\; 3 - i) \in \mathbb{C}^2,
$$

then

$$
\text{Re}\{\vec{v}\} = (1, 3), \quad \text{Im}\{\vec{v}\} = (2, -1).
$$

```python
import numpy as np

vector_v = np.array([1 + 2j, 3 - 1j])
vector_w = np.array([2 - 1j, 4 + 3j])

print("v =", vector_v)
print("Re{v} =", vector_v.real)
print("Im{v} =", vector_v.imag)

vector_sum = vector_v + vector_w
print("\nv + w =", vector_sum)

scalar = 2 + 1j
scaled_vector = scalar * vector_v
print("\n(2+i)¬∑v =", scaled_vector)
```
