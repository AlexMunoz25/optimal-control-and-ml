### 1.1.5.1.1. Lines and Planes

$$
\ell: \{p_o + t\,\vec{v},\; t \in \mathbb{R}\}
$$

$$
P: \{Ax + By + Cz = D\}
$$

$$
P: \{\vec{n} \cdot [(x,y,z) - p_o] = 0\}
$$

**Explanation:**

Points, lines, and planes are the basic building blocks of geometry. A point in $\mathbb{R}^3$ is specified by its coordinates $p = (p_x, p_y, p_z)$. A line $\ell$ is a one-dimensional space described by a direction vector and a point. A plane $P$ is a two-dimensional space described by a normal vector and a point.

A point can also be specified as the intersection of two lines (in $\mathbb{R}^2$) or three planes (in $\mathbb{R}^3$), which corresponds to solving systems of linear equations.

**Example:**

Find where the lines $x + 2y = 5$ and $3x + 9y = 21$ intersect.

Solving simultaneously: $p = (1, 2)$.

```python
import numpy as np

coefficient_matrix = np.array([[1, 2], [3, 9]])
constants = np.array([5, 21])

intersection = np.linalg.solve(coefficient_matrix, constants)
print(f"Intersection point: ({intersection[0]}, {intersection[1]})")
```
