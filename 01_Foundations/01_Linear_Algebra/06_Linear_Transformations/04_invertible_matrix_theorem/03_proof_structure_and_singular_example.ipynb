### 1.1.6.4.3. Proof Structure and Singular Example

$$
(1) \Rightarrow (2) \Rightarrow (3) \Rightarrow (4) \Rightarrow (5) \Rightarrow (6) \Rightarrow (7) \Rightarrow (1)
$$

**Explanation:**

The proof establishes a closed chain of implications between statements (1) through (7), with additional links $(1) \Leftrightarrow (8) \Leftrightarrow (9)$ and $(5) \Leftrightarrow (10)$.

Key implications:
- $(1) \Rightarrow (2)$: If $A^{-1}$ exists, $\vec{x} = A^{-1}\vec{b}$ is the unique solution
- $(2) \Rightarrow (3)$: Unique solution implies trivial null space (otherwise $\vec{x} + \vec{y}$ with $\vec{y} \in \mathcal{N}(A)$ would be a second solution)
- $(4) \Rightarrow (5)$: Only trivial solution to $A\vec{v} = \vec{0}$ means columns are linearly independent
- $(7) \Rightarrow (1)$: $\operatorname{rref}(A) = \mathbb{1}_n$ implies $A^{-1} = E_k \cdots E_2 E_1$ (product of elementary matrices)
- $(1) \Leftrightarrow (8)$: $(A^{-1})^\mathsf{T} A^\mathsf{T} = \mathbb{1}_n$, so $A^\mathsf{T}$ is invertible iff $A$ is invertible

**Example (singular matrix):**

$B = \begin{bmatrix} 1 & 2 \\ 2 & 4 \end{bmatrix}$: $\det(B) = 0$, $\operatorname{rank}(B) = 1 < 2$, $\mathcal{N}(B) = \operatorname{span}\{(-2, 1)\}$ â€” all 10 statements are FALSE.

```python
import sympy as sp

matrix_b = sp.Matrix([[1, 2], [2, 4]])

print("B =")
sp.pprint(matrix_b)
print(f"det(B) = {matrix_b.det()}")
print(f"rank(B) = {matrix_b.rank()}")

rref_b, _ = matrix_b.rref()
print("rref(B) =")
sp.pprint(rref_b)

null_b = matrix_b.nullspace()
print(f"N(B) = span{list(null_b[0])}")
```
