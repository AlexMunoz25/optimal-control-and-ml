### 1.1.6.5. P6.2 ‚Äî Transformation on Function Spaces

$$
V = \left\{ \alpha\, e^{2x}\cos(x) + \beta\, e^{2x}\sin(x) \;\middle|\; \alpha, \beta \in \mathbb{R} \right\}
$$

$$
L : V \to V, \quad L(f) = f' + f
$$

$$
M_L = \begin{bmatrix} 3 & 1 \\ -1 & 3 \end{bmatrix}
$$

**Explanation:**

The vector space $V$ is spanned by $\{e^{2x}\cos(x),\; e^{2x}\sin(x)\}$. The linear operator $L(f) = f' + f$ maps $V$ into itself. To find $M_L$, probe with each basis function and express the result in the same basis.

**Example:**

Let $f_1 = e^{2x}\cos(x)$ and $f_2 = e^{2x}\sin(x)$.

$$
f_1' = 2e^{2x}\cos(x) - e^{2x}\sin(x)
$$

$$
L(f_1) = f_1' + f_1 = 3e^{2x}\cos(x) - e^{2x}\sin(x) = 3f_1 - f_2
$$

$$
f_2' = 2e^{2x}\sin(x) + e^{2x}\cos(x)
$$

$$
L(f_2) = f_2' + f_2 = e^{2x}\cos(x) + 3e^{2x}\sin(x) = f_1 + 3f_2
$$

$$
M_L = \begin{bmatrix} 3 & 1 \\ -1 & 3 \end{bmatrix}
$$

```python
import sympy as sp

x_sym = sp.Symbol("x")

basis_f1 = sp.exp(2 * x_sym) * sp.cos(x_sym)
basis_f2 = sp.exp(2 * x_sym) * sp.sin(x_sym)

l_f1 = sp.diff(basis_f1, x_sym) + basis_f1
l_f2 = sp.diff(basis_f2, x_sym) + basis_f2

l_f1_simplified = sp.expand(l_f1)
l_f2_simplified = sp.expand(l_f2)

print(f"L(f1) = {l_f1_simplified}")
print(f"L(f2) = {l_f2_simplified}")

coefficients_f1 = [l_f1_simplified.coeff(basis_f1), l_f1_simplified.coeff(basis_f2)]
coefficients_f2 = [l_f2_simplified.coeff(basis_f1), l_f2_simplified.coeff(basis_f2)]

matrix_l = sp.Matrix([[coefficients_f1[0], coefficients_f2[0]],
                       [coefficients_f1[1], coefficients_f2[1]]])

print(f"\nM_L =\n{matrix_l}")
```
