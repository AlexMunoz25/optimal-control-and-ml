### 1.1.1.4. Euclidean Space

$$
X = \sum_{i=1}^{n} \lambda_i V_i
$$

$$
\vec{e}_3 = \vec{e}_1 \times \vec{e}_2
$$

**Explanation:**

A Euclidean space is a linear vector space where all points can be represented as linear combinations of mutually perpendicular basis vectors.
All points are expressed as $X = \sum_{i=1}^{n} \lambda_i V_i$ where $\lambda_i$ are scalars and $V_i$ are basis vectors.

The distance of a point $\vec{P} = (x, y, z)$ from the origin is $\sqrt{x^2 + y^2 + z^2}$.

The Cartesian coordinate system specifies directions using the right-hand rule: if the thumb points along $\vec{e}_1$ and the index finger along $\vec{e}_2$, the middle finger gives $\vec{e}_3$.

**Example:**

For point $P = (3, 4, 12)$, the distance from the origin is $\sqrt{9 + 16 + 144} = \sqrt{169} = 13$.

```python
import numpy as np

unit_e1 = np.array([1, 0, 0])
unit_e2 = np.array([0, 1, 0])
unit_e3 = np.array([0, 0, 1])

point = np.array([3, 4, 12])

distance = np.linalg.norm(point)

cross_product = np.cross(unit_e1, unit_e2)

print("Distance from origin:", distance)
print("Right-hand rule (e1 √ó e2):", cross_product)
```
