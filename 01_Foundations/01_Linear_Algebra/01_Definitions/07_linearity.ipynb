### 1.1.1.7 Linearity


$$
f(x) =
\underbrace{\frac{a}{x}}_{\text{one-over-x}}
+ \underbrace{b}_{\text{constant}}
+ \underbrace{mx}_{\text{linear term}}
+ \underbrace{qx^2}_{\text{quadratic}}
+ \underbrace{cx^3}_{\text{cubic}}.
$$


**Explanation:**

The term $mx$ is called the **linear term** because it contains $x$ raised to the first power.  
All other terms ‚Äî $\frac{a}{x}$, $qx^2$, and $cx^3$ ‚Äî are **nonlinear**, since they involve $x$ either in the denominator or raised to powers other than one.  

The **linear term** is unique because changes in $x$ lead to *proportional* changes in the output.  
If $x$ doubles, the linear term doubles: $m(2x) = 2mx$.  
If $x$ increases by a factor of 100, the term becomes $100mx$.  
This proportionality does not hold for nonlinear terms, which change at non-constant rates relative to $x$.

---

#### Introduction

A **single-variable function** takes a real number $x$ as input and outputs another real number $f(x)$.  
This relationship is written as:

$$
f: \mathbf{R} \rightarrow \mathbf{R}
$$

The function $f$ maps each real input to exactly one real output.  
In studying linearity, we focus on functions or equations that contain **only linear terms**, where $x$ appears with power one.

---

**Example:**

For example, consider the linear function:

$$
f(x) = 3x + 2
$$

If $x = 1$, then $f(1) = 5$.  
If $x = 2$, then $f(2) = 8$.  
Doubling $x$ doubles the $3x$ term ‚Äî showing perfect linear proportionality.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
f_linear = 3*x + 2
f_nonlinear = 3*x**2 + 2

plt.plot(x, f_linear, label="Linear: f(x)=3x+2")
plt.plot(x, f_nonlinear, "--", label="Nonlinear: f(x)=3x¬≤+2")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Comparison: Linear vs Nonlinear Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
