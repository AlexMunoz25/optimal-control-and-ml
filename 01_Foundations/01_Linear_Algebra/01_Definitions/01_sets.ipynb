Mathematicians classify the different kinds of number-like objects into categories called **sets**:

- **The natural numbers:**  
  $\mathbb{N} = \{0, 1, 2, 3, 4, 5, 6, 7, \ldots\}$

- **The integers:**  
  $\mathbb{Z} = \{\ldots, -3, -2, -1, 0, 1, 2, 3, \ldots\}$

- **The rational numbers:**  
  $\mathbb{Q} = \left\{\frac{5}{3}, \frac{22}{7}, 1.5, 0.125, -7, \ldots\right\}$

- **The real numbers:**  
  $\mathbb{R} = \{-1, 0, 1, \sqrt{2}, e, \pi, 4.94\ldots, \ldots\}$

- **The complex numbers:**  
  $\mathbb{C} = \{-1, 0, 1, i, 1+i, 2+3i, \ldots\}$


**Definitions:**

- *Set*: a collection of mathematical objects.  
- $S, T$: the usual variable names for sets.  
- $s \in S$: this statement is read ‚Äú$s$ is an element of $S$‚Äù or ‚Äú$s$ is in $S$.‚Äù  
- $\mathbb{N}, \mathbb{Z}, \mathbb{Q}, \mathbb{R}$: important number sets ‚Äî the naturals, the integers, the rationals, and the reals, respectively.  
- $\varnothing$: the *empty set* is a set that contains no elements.  
- $\{\, \cdot \,\}$: the curly brackets are used to define sets, and the expression inside the curly brackets describes the set contents.


**Set operations:**

- $S \cup T$: the *union* of two sets. The union of $S$ and $T$ corresponds to the elements in either $S$ or $T$.  
- $S \cap T$: the *intersection* of the two sets. The intersection of $S$ and $T$ corresponds to the elements that are in both $S$ and $T$.  
- $S \setminus T$: the *set difference* or *set minus*. The set difference $S \setminus T$ corresponds to the elements of $S$ that are not in $T$.


**Set relations:**

- $S \subset T$: is a *strict subset* of.  
- $S \subseteq T$: is a *subset of or equal to*.


**Example:**

Let  

$$
S = \{1, 2, 3\}, \quad T = \{3, 4\}.
$$  

Then  

$$
S \cup T = \{1, 2, 3, 4\}, \quad S \cap T = \{3\}, \quad S \setminus T = \{1, 2\}.
$$





```python
S = {1, 2, 3}
T = {3, 4}

print("S =", S)
print("T =", T)
print("S ‚à™ T =", S | T)         # Union
print("S ‚à© T =", S & T)         # Intersection
print("S \\ T =", S - T)        # Difference (elements in S not in T)
print("T \\ S =", T - S)        # Difference (elements in T not in S)
print("S ‚ñ≥ T =", S ^ T)         # Symmetric difference (elements in either S or T, but not both)
print("S ‚äÇ T =", S < T)         # Strict subset
print("S ‚äÜ T =", S <= T)        # Subset or equal
print("S ‚äÉ T =", S > T)         # Strict superset
print("S ‚äá T =", S >= T)        # Superset or equal
print("S ‚à© T = ‚àÖ:", S.isdisjoint(T))  # Disjoint check
```
